```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

# Load Data

```python
df = pd.read_csv("../data/sa1_prices_2024.csv", header=0, names=["datetime", "region", "rrp"])
df["datetime"] = pd.to_datetime(df["datetime"])
df["rrp"] = pd.to_numeric(df["rrp"], errors="coerce")
df = df.dropna()
```

```python
df["negative"] = (df["rrp"] < 0).astype(int)
df["hour"] = df["datetime"].dt.hour
df["dayofweek"] = df["datetime"].dt.dayofweek
df["month"] = df["datetime"].dt.month
df["lag_1"] = df["rrp"].shift(1)
df["lag_6"] = df["rrp"].shift(6)
df["lag_12"] = df["rrp"].shift(12)
df = df.dropna()
```

```python
features = ["hour", "dayofweek", "month", "lag_1", "lag_6", "lag_12"]
X_train, X_test, y_train, y_test = train_test_split(
    df[features], df["negative"], test_size=0.2, shuffle=False
)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
print(classification_report(y_test, y_pred := rf.predict(X_test)))
```
