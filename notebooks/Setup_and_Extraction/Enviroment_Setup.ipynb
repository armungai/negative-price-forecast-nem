# Part 1 — Create the virtual environment


This section creates a virtual environment **two folders above this notebook** at:

`../../.venv`

### What this part does
1. Shows the current Python executable
2. Creates the virtual environment at `../../.venv`
3. Finds the Python executable inside that environment
4. Upgrades `pip`, `setuptools`, and `wheel`
5. Installs `ipykernel`
6. Registers the environment as a Jupyter kernel

### Important
After running all cells in Part 1, go to:

**Kernel > Change Kernel > Python (../../.venv)**

Then continue with **Part 2**.

```python
from pathlib import Path
import sys

venv_path = Path("../../.venv").resolve()

print("Current Python executable:")
print(sys.executable)
print()
print("Virtual environment will be created at:")
print(venv_path)
```

## Step 1 — Create the virtual environment directory

This cell creates the virtual environment at `../../.venv`.

If it already exists, it will not recreate it.

```python
import subprocess
from pathlib import Path
import sys

venv_path = Path("../../.venv").resolve()

if venv_path.exists():
    print(f"Virtual environment already exists at: {venv_path}")
else:
    print(f"Creating virtual environment at: {venv_path}")
    subprocess.run([sys.executable, "-m", "venv", str(venv_path)], check=True)
    print("Virtual environment created successfully.")
```

## Step 2 — Find the Python executable inside the virtual environment

The location is different on Windows vs macOS/Linux, so this cell detects it automatically.

```python
import os
from pathlib import Path

venv_path = Path("../../.venv").resolve()

if os.name == "nt":
    venv_python = venv_path / "Scripts" / "python.exe"
else:
    venv_python = venv_path / "bin" / "python"

print("Virtual environment Python executable:")
print(venv_python)
print()
print("Exists:", venv_python.exists())
```

## Step 3 — Upgrade packaging tools inside the virtual environment

This upgrades:
- `pip`
- `setuptools`
- `wheel`

```python
import subprocess

print("Upgrading pip, setuptools, and wheel inside ../../.venv ...")
subprocess.run(
    [str(venv_python), "-m", "pip", "install", "--upgrade", "pip", "setuptools", "wheel"],
    check=True
)
print("Upgrade complete.")
```

## Step 4 — Install `ipykernel` into the virtual environment

This is required so Jupyter can use the new environment as a notebook kernel.

```python
import subprocess

print("Installing ipykernel inside ../../.venv ...")
subprocess.run(
    [str(venv_python), "-m", "pip", "install", "ipykernel"],
    check=True
)
print("ipykernel installed.")
```

## Step 5 — Register the virtual environment as a Jupyter kernel

After this, you will be able to switch the notebook to:

**Python (../../.venv)**

```python
import subprocess

print("Registering kernel: Python (../../.venv)")
subprocess.run([
    str(venv_python),
    "-m",
    "ipykernel",
    "install",
    "--user",
    "--name=venv_2up",
    "--display-name=Python (../../.venv)"
], check=True)

print("Kernel registered successfully.")
```

## Step 6 — Switch the notebook kernel

Now do the following manually:

1. Open the notebook menu
2. Click **Kernel**
3. Click **Change Kernel**
4. Select **Python (../../.venv)**

Only after switching kernels should you continue to Part 2.

## Step 2 — Print the Python version and installed package versions

```python
import sys
import platform
from importlib.metadata import version, PackageNotFoundError

packages = [
    "nemosis",
    "pandas",
    "numpy",
    "matplotlib",
    "seaborn",
    "scipy",
    "statsmodels",
    "scikit-learn",
    "jupyterlab",
    "ipykernel",
    "pip",
    "setuptools",
    "wheel",
]

print("Python Version Information")
print("-" * 50)
print("Python version :", platform.python_version())
print("Executable     :", sys.executable)
print("Platform       :", platform.platform())
print()

print("Installed Package Versions")
print("-" * 50)
for pkg in packages:
    try:
        print(f"{pkg:<15} : {version(pkg)}")
    except PackageNotFoundError:
        print(f"{pkg:<15} : NOT INSTALLED")
```
