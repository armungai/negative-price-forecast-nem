# For Windows users

Set the paths to your R installation location

```python
import os
import sys

# 1. Point to your base R installation folder
os.environ['R_HOME'] = r'C:\Program Files\R\R-4.6.0'

# 2. Add the core R binaries and the exact folder holding stats.dll to your system PATH
r_bin_path = r'C:\Program Files\R\R-4.6.0\bin\x64'
r_stats_path = r'C:\Program Files\R\R-4.6.0\library\stats\libs\x64'

# Prepend them to the environment PATH variable
os.environ['PATH'] = f"{r_bin_path};{r_stats_path};{os.environ['PATH']}"

# 3. Now load rpy2 cleanly
%load_ext rpy2.ipython
print("Environment variables patched successfully!")
```
